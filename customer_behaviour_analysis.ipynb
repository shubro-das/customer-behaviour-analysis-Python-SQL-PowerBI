# Customer shopping behavior analysis

```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
```

```python
df.columns
```

### Create a column age_group

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head()
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
frequency_mapping = {
    'Every 3 Months': 90,
    'Annually': 365,
    'Quarterly': 90,
    'Monthly': 30,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Weekly': 7
}
```

### Create a columns purchase_frequency_days

```python

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head()
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# checking if discount applied is the same as promo code used
(df['discount_applied'] == df['promo_code_used']).all()
```

#### Let's drop one of the redundant columns

```python
df = df.drop('promo_code_used', axis=1)
```
